```python
def addNeighbours(array, k):
        neighbours = []
        array = sorted(array)
        neighbours.append(array[:k])
        return neighbours
        
```

```python
def KNN (test, training, k):
    if k > len(training):
        return 'k is larger than set'
    elif k<1:
        return 'please make sure that k>0'
    else:
        array = []
        arrayClass = []
        neighbours = []
        for i in range(len(test)):
            distance = []
            Distance_class = []
            for j in range(len(training)):
                value = 0.0
                for m in range(len(test[0])-1):
                    value += (np.abs(((test[i][m] - training[j][m]))))
                distance.append(value)
                Distance_class.append([value , training[j][-1]])
            array.append(distance)
            arrayClass.append(Distance_class)
            neighbours += (addNeighbours(array[i], k))
        arrayClass = np.array(arrayClass)
        neighbours = np.array(neighbours)
        # array classes is 3d
        # the big list holding all distance,class for all test points to all training points
        # smaller lists holding multiple distance,class for one test point and multiple training point
        # smallest one holding a distance, class between one test point and one training point
        return finalize(test, training, arrayClass, neighbours)
```

```python
def finalize ( test, training, array, neighbours):
    predicted = [] #holds smallest k distances in output list
    expected = []
    output = [] #holds all outputs
    appended = []
    for i in range(len(neighbours)):
        classes=[]
        for j in range(len(neighbours[0])):
            for m in range((len(array[0])-1)):
                if float(neighbours[i][j]) == float(array[i][m][:-1]):
                    if (i,j) not in appended:
                        appended.append((i,j))
                        classes.append(array[i][m][-1])
        expected.append(test[i][-1])
        output.append(classes)
        
    output = np.array(output, dtype=object)
    for i in output:
        predicted.append(Count(i))
        
    dic= {
        'Exact Output':expected,
        'Predicted Output':predicted}    
    table = pd. DataFrame(dic)
    return table
    
```

```python
def Count(output):
    count = {}
    for variety in output:
            if variety not in count:
                count [variety] = 1
            else:
                 count[variety] += 1
    return predict(count)
```

```python
def predict(output):
    maximum = 0.0
    classifier = ''
    for value in output.values():
        if value > maximum:
            maximum = value
    for key in output:
        if output[key] == maximum:
            classifier = key
    return classifier
```
